```python
import duckdb
import pandas as pd

con = duckdb.connect()
print("DuckDB conectado!")
```

Visão geral da camada silver

```python
df = con.execute("""
    SELECT
        start_time,
        distance_km,
        avg_pace_min_km / 60 AS pace_min_km,
        duration_min,
        calories,
        subjective_feeling,
        weather_condition,
        temperature_c
    FROM read_parquet('data/silver/runs.parquet')
    ORDER BY start_time
    LIMIT 10
""").df()

df
```

Resumo anual

```python
con.execute("""
    SELECT *
    FROM read_parquet('data/gold/yearly_summary.parquet')
    ORDER BY year
""").df()
```

Evolução mensal de km

```python
con.execute("""
    SELECT
        year_month,
        total_runs,
        total_km,
        avg_pace_min_km / 60 AS avg_pace_min_km
    FROM read_parquet('data/gold/monthly_summary.parquet')
    ORDER BY year_month
""").df()
```
